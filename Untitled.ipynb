```python
for i in range(21):
    df.iloc[i] = df.iloc[i].str.replace(')', 'c')
```

```python
for i in range(21):
    df.iloc[i] = df.iloc[i].str.replace('}', 'd')
```

```python
for i in range(21):
    df.iloc[i] = df.iloc[i].str.replace('{', 'e')
```
